# Matplotlib

Paquete de python usado para crear gráficos, imágenes y animaciones.

````python
#Importamos el paquete
import matplotlib.pyplot as plt  #Así se hace por convención
````

Así como la clase ```array```es propia de la librería numpy, ```figure``` es una clase propia de matplotlib. 

Matplotlib grafica nuestros datos en objetos de clase ```figure```, cada objeto contiene una o más ```axes```

````python
fig, ax = plt.subplots() #Crea un figure que contiene un axes

#Ploteamos una relación de datos (x,y)
ax.plot([1,2,3,4],[1,4,2,3])

#ploteamos otra relación de datos (x,y)
ax.plot([1,2,3,4],[2.5,2.5,2.5,2.5])
````

# ```Figure```

````python
#Creamos un objeto de clase figure
#Esta figura está vacía y no tiene Axes

fig = plt.figure(figsize = (5,5), dpi = 10)  

#Figsize --> Altura y anchura en pulgadas
#dpi --> Resolución en puntos por pulgadas
#facecolor --> Color del fondo
#Edgecolor --> Color del borde
````

````python
#O podemos crear la figura y sus Axes al mismo tiempo
fig,ax = plt.subplots()
#Esta figura tiene un solo Axes
````

````python
#Sin embargo, podemos crear figures con más Axes
fig,ax = plt.subplots(1,2, figsize = (7,3))
````

# ```Axes```

Los axes son ```Artist``` adjuntos a los objetos clase figure, cada axes tiene 2 o 3 ```Axis```, que básicamente son los ejes.

````python
ax[0].set_title('Podemos agregar titulos')
ax[0].set_xlabel('Podemos agregar etiquetas')
ax[1].set_ylabel('En todos los ejes')
ax[1].set_xlabel('De todos los Axes')

import numpy as np
x = np.linspace(0,1,100)
y = np.cos(10*x)/2
y2 = y/5

ax[0].plot(x,y, 'go--', linewidth = 2, markersize = 3)
ax[0].plot(x,y2, color = 'orange', marker = 'o')
ax[1].plot(x,y,'+')
ax[1].plot(x,y2)

ax[0].grid()
ax[1].grid()

fig
````
